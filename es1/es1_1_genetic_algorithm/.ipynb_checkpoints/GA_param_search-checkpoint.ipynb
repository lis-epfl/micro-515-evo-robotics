## GA parameters
The default parameters provided here are by no means optimal. You should try different values to see what happens. However, to understand how changing one of the parameters influences the search for a solution, you should look at the mean and standard deviation of the fitness to see how the rate of convergence and best solution found are influenced. For instance, if you tried mutation probabilities of 0.01, 0.1, 0.2, 0.3, and 0.4, ran each 100 generations and repeated this 10 times each, then you could plot the mean and standard deviation of the fitness against the number of generations to find a good mutation rate that results in finding the best x and y parameters in the lowest number of generations on average.

It is often a good idea to keep track of the best solution thus far so that it is not lost. If ELITISM=True, then this best solution will always be added to the new population without being mutated so that future generations can continue to search this area of the solution space.

```python
# GA parameters
NUM_GENERATIONS = 100
POPULATION_SIZE = 100
TOURNAMENT_SIZE = 3
PROB_MUTATION = 0.04
PROB_CROSSOVER = 0.5
STAT_INTERVALS = 10

# add the best chromosome in the current generation to the next generation
ELITISM = True

# we will add the best individual to the population if ELITISM
POPULATION_SIZE = POPULATION_SIZE - 1 if ELITISM else POPULATION_SIZE
```

## Statistical study of GA parameters

First, you should choose which parameter you want to vary and which values of this parameter you want to try. Here, we have selected PROB_MUTATION and the values of PROB_MUTATION that we want to test are 0.01, 0.1, 0.2, 0.3, and 0.4. The best value may not be any of these though so you should use the results you observe to update the searched values.

You should also select how many times you want to repeat the GA for each chosen value (REPETITIONS). The results will be plotted at the end so you can try to visually see which mutation probability is, on average, likely to result in a faster convergence.

```python
PARAM_TYPE = "PROB_MUTATION"
PARAM_OPTIONS = [0.01, 0.1, 0.2, 0.3, 0.4] # e.g. 5 different mutation probs

REPETITIONS = 10
```

Note: remember to rename the first *for* loop if PARAM_TYPE is not "PROB_MUTATION"

```python
# save statistics for each parameter option
all_options_mean = []
all_options_std = []

for po in PARAM_OPTIONS:
    exec(f"{PARAM_TYPE}={po}")
    print(f"Starting {PARAM_TYPE}={po}")
    option_best = []
    for _ in range(REPETITIONS):

        # initial population 
        chromosome = np.array([0,0,1,0,1,0,1,1,1,1,1,1,0,1,1,1,0,0,1,0])
        population = np.empty((0,len(chromosome)))
        for i in range(POPULATION_SIZE):
            random.shuffle(chromosome)
            population = np.vstack((population, chromosome))

        # keep track of the best chromosome
        best_chromosome = np.empty((0,len(chromosome)))
        best_fitness = 100000
        repetition_best = []

        for i in range(NUM_GENERATIONS + 1):

            # evaluate the fitness for each chromosome in the current generation
            x_y_fit = np.array([ackley(chrsm) for chrsm in population])
            
            gen_best_idx = np.argmin(x_y_fit[:, 2])
            gen_best_x, gen_best_y, gen_best_fit = x_y_fit[gen_best_idx, 0], x_y_fit[gen_best_idx, 1], x_y_fit[gen_best_idx, 2]

            # save information about the current generation
            repetition_best.append(gen_best_fit)

            next_gen = np.empty((0,len(chromosome)))
            while len(next_gen) < POPULATION_SIZE:
                # select parent chromosomes
                parent_1 = tournament_selection(population, x_y_fit[:, 2], TOURNAMENT_SIZE)
                parent_2 = tournament_selection(population, x_y_fit[:, 2], TOURNAMENT_SIZE)
                # make child chromosomes using 2-point crossover
                children = crossover(parent_1, parent_2, PROB_CROSSOVER)
                next_gen = np.vstack((next_gen, children[0]))
                if len(next_gen) >= POPULATION_SIZE: break
                next_gen = np.vstack((next_gen, children[1]))

            # mutate genes in the next generation
            next_gen = mutation(next_gen, PROB_MUTATION)

            if ELITISM:
                if gen_best_fit < best_fitness:
                    best_fitness = gen_best_fit
                    best_chromosome = population[gen_best_idx]
                # add the best chromosome to the next generation without mutation
                next_gen = np.vstack((next_gen, best_chromosome))

            population = next_gen.copy()
            
        option_best.append(repetition_best)

    option_best = np.array(option_best)
    
    m = np.mean(option_best, axis=0)
    s = np.std(option_best, axis=0)
    
    all_options_mean.append(m)
    all_options_std.append(s)

fig, ax = plt.subplots()

x_axis_range = list(range(m.shape[0]))

COLOURS = ['b', 'c', 'g', 'm', 'y', 'r']

ax.plot(x_axis_range, np.zeros(len(x_axis_range)), ':r', linewidth=1.5)

for i in range(len(PARAM_OPTIONS)):
    ax.plot(x_axis_range, all_options_mean[i], '--'+COLOURS[i])
    ax.fill_between(x_axis_range, all_options_mean[i]-all_options_std[i], all_options_mean[i]+all_options_std[i], color = COLOURS[i], alpha=0.1)

ax.set_ylabel("Fitness")
ax.set_xlabel("Generations")

patches = []
for i in range(len(PARAM_OPTIONS)):
    patches.append(mpatches.Patch(color=COLOURS[i], label=f"{PARAM_TYPE}: {PARAM_OPTIONS[i]}"))

ax.legend(handles=patches)

plt.show()
```
